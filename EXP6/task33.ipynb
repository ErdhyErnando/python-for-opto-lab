```python
import matplotlib.pyplot as plt

# Data
current = [20.00, 21.03, 22.04, 23.03, 24.01, 25.03, 26.03]  # Diode current in mA
delta_lambda = [0, 0.86e14, 4.70e14, 8.74e14, 1.22e15, 1.68e15, 1.95e15]  # Δλ in nm

# Create the plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.plot(current, delta_lambda, marker='o', linestyle='-', color='b', label='Δλ vs Current')

# Add labels and title
plt.xlabel('Diode Current (mA)', fontsize=12)
plt.ylabel('Δλ (nm)', fontsize=12)
plt.title('Wavelength Shift (Δλ) vs Diode Current', fontsize=14)

# Add grid
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
