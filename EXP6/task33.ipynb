```python
import matplotlib.pyplot as plt

# Data
current = [20.00, 21.03, 22.04, 23.03, 24.01, 25.03, 26.03]  # Diode current in mA
delta_lambda = [0, 0.86e14, 4.70e14, 8.74e14, 1.22e15, 1.68e15, 1.95e15]  #  in nm

# Create the plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.plot(current, delta_lambda, marker='o', linestyle='-', color='b', label=' vs Current')

# Add labels and title
plt.xlabel('Diode Current (mA)', fontsize=12)
plt.ylabel(' (nm)', fontsize=12)
plt.title('Wavelength Shift () vs Diode Current', fontsize=14)

# Add grid
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_with_slope(curr, delta_lambda):
    # Convert lists to numpy arrays for easier calculation
    curr = np.array(curr)
    delta_lambda = np.array(delta_lambda)
    
    # Calculate the slope using numpy's polyfit
    # polyfit returns coefficients of the line of best fit (slope is the first coefficient)
    slope, intercept = np.polyfit(curr, delta_lambda, 1)
    
    # Create the line of best fit
    line_of_best_fit = slope * curr + intercept
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    
    # Plot original data points
    plt.plot(curr, delta_lambda, 'o', label='Data Points')
    
    # Plot line of best fit
    plt.plot(curr, line_of_best_fit, 'r--', label=f'Slope = {slope:.2e} nm/mA')
    
    # Add labels and title
    plt.xlabel('Diode Current (mA)', fontsize=12)
    plt.ylabel(' (nm)', fontsize=12)
    plt.title('Wavelength Shift () vs Diode Current with Slope', fontsize=14)
    
    # Add grid
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # Add legend
    plt.legend()
    
    # Show the plot
    plt.show()
    
    return slope

# Data
current = [20.00, 21.03, 22.04, 23.03, 24.01, 25.03]  # Diode current in mA
delta_lambda = [1.55E-14, 2.93E-14, 4.40E-14, 5.65E-14, 7.28E-14, 8.28E-14]  #  in nm

# Call the function and get the slope
slope = plot_with_slope(current, delta_lambda)
print(f"\nThe slope of the line is: {slope:.2e} nm/mA")
```
