```python
import matplotlib.pyplot as plt

# Data
temperature = [20.0, 20.1, 20.2, 20.3, 20.4, 20.5, 20.6, 20.7, 20.8]  # Temperature in °C
delta_lambda = [0, 0, 1.56e14, 4.44e14, 7.44e14, 1.07e15, 1.22e15, 1.55e15, 1.84e15]  # Δλ in nm

# Create the plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.plot(temperature, delta_lambda, marker='o', linestyle='-', color='b', label='Δλ vs Temperature')

# Add labels and title
plt.xlabel('Temperature (°C)', fontsize=12)
plt.ylabel('Δλ (nm)', fontsize=12)
plt.title('Wavelength Change (Δλ) vs Temperature', fontsize=14)

# Add grid
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
