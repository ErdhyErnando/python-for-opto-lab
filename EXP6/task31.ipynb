```python
import matplotlib.pyplot as plt
import numpy as np

def plot_with_slope(temp, delta_lambda):
    # Convert lists to numpy arrays for easier calculation
    temp = np.array(temp)
    delta_lambda = np.array(delta_lambda)
    
    # Calculate the slope using numpy's polyfit
    # polyfit returns coefficients of the line of best fit (slope is the first coefficient)
    slope, intercept = np.polyfit(temp, delta_lambda, 1)
    
    # Create the line of best fit
    line_of_best_fit = slope * temp + intercept
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    
    # Plot original data points
    plt.plot(temp, delta_lambda, 'o', label='Data Points')
    
    # Plot line of best fit
    plt.plot(temp, line_of_best_fit, 'r--', label=f'Slope = {slope:.2e} pm/°C')
    
    # Add labels and title
    plt.xlabel('Temperature (°C)', fontsize=12)
    plt.ylabel('Δλ (pm)', fontsize=12)
    plt.title('Wavelength Change (Δλ) vs Temperature with Slope', fontsize=14)
    
    # Add grid
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # Add legend
    plt.legend()
    
    # Show the plot
    plt.show()
    
    return slope

# Data
temperature = [20.0, 20.1, 20.2, 20.3, 20.4, 20.5, 20.6, 20.7]  # Temperature in °C
delta_lambda = [1.21E-12, 1.77E-12, 2.80E-12, 3.88E-12, 5.04E-12, 5.60E-12, 6.77E-12, 7.80E-12]  # Δλ in pm

# Call the function and get the slope
slope = plot_with_slope(temperature, delta_lambda)
print(f"\nThe slope of the line is: {slope:.2e} pm/°C")
```
