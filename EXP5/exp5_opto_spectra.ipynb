### This jupyter is for parsing and plotting the spectra `.ssm` data

```python
import re
import os
import matplotlib.pyplot as plt
```

```python
directory = "gruppe 7 WS24"
files = os.listdir(directory)
print(files)
```

```python
def parse_ssm_file(file_path):
    x_values = []
    y_values = []

    with open(file_path, 'r') as file:
        for line in file:

            # skip header
            if "File" in line or "SCOPE->" in line:
                continue

            # extract x and y values using regex
            matches = re.findall(r"(\d+\.\d+)\s+([+-]?\d+\.\d+[Ee][+-]?\d+)", line)

            for match in matches:
                x_values.append(float(match[0]))
                y_values.append(float(match[1]))

    return x_values, y_values
```

```python
# Usage
# file_path =  'gruppe 7 WS24/20after.ssm'
# file_path =  'gruppe 7 WS24/20before.ssm'
# file_path =  'gruppe 7 WS24/25before.ssm'
# file_path =  'gruppe 7 WS24/25after.ssm'
# file_path =  'gruppe 7 WS24/30before.ssm'
# file_path =  'gruppe 7 WS24/30after.ssm'
# file_path =  'gruppe 7 WS24/35before.ssm'
# file_path =  'gruppe 7 WS24/35after.ssm'
file_path =  'gruppe 7 WS24/40after.ssm'
# file_path =  'gruppe 7 WS24/40before.ssm'
x_axis, y_axis = parse_ssm_file(file_path)


# print value
print("x-axis:", x_axis)
print("y-axis:", y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```
